```python
#Gereksiz Sütunlar
df.drop(columns=['Unnamed: 0', 'Uri', 'Url_spotify', 'Url_youtube', 'Description', 'Licensed', 'official_video', ], inplace=True)
```

```python
df.head()
```

```python
# Null değerleri kontrol edeceğimiz sütunlar
columns_to_check = [
    'Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness', 
    'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 
    'Tempo', 'Duration_ms'

df.dropna(subset=columns_to_check, inplace=True)
df.isnull().sum()
```

```python
#Eksik değerleri doldurma 
df['Views'] = df['Views'].fillna(df['Views'].mean())
df['Likes'] = df['Likes'].fillna(df['Likes'].mean())
df['Comments'] = df['Comments'].fillna(df['Comments'].mean())
df['Stream'] = df['Stream'].fillna(df['Stream'].mean())
df['Title'] = df['Title'].fillna('Unknown Title')
df['Channel'] = df['Channel'].fillna('Unknown Channel')

```

```python
df.isnull().sum()
```

<h2>Bir şarkı veya video analiz edilirken, izlenme sayısı, yorum sayısı ve beğeni sayıları en önemli üç kriter olarak öne çıkmaktadır.
Bu kriterler göz önünde bulundurularak yapılan bir analizde, söz konusu sütunlarda yer alan aykırı değerlerin belirlenmesi 
ve değerlendirilmesi büyük önem taşımaktadır.<h2/>

```python
plt.figure(figsize=(14, 6))

# Views için box-plot
plt.subplot(1, 3, 1)
sns.boxplot(y=df['Views'])
plt.title('Views')

# Likes için box-plot
plt.subplot(1, 3, 2)
sns.boxplot(y=df['Likes'])
plt.title('Likes')

# Comments için box-plot
plt.subplot(1, 3, 3)
sns.boxplot(y=df['Comments'])
plt.title('Comments')

plt.tight_layout()
plt.show()
```

```python
# Views için aykırı değer tespiti 
Q1_views = df['Views'].quantile(0.25)  
Q3_views = df['Views'].quantile(0.75)  
IQR_views = Q3_views - Q1_views

lower_bound = Q1 - 1.5 * IQR_views
upper_bound = Q3 + 1.5 * IQR_views

df_cleaned_views = df[(df['Views'] >= lower_bound) & (df['Views'] <= upper_bound)]
df_cleaned_views.shape
```

```python
#Likes için aykırı değer tespiti
Q1_likes = df['Likes'].quantile(0.25) 
Q3_likes = df['Likes'].quantile(0.75) 
IQR_likes = Q3_likes - Q1_likes  

lower_bound_likes = Q1_likes - 1.5 * IQR_likes
upper_bound_likes = Q3_likes + 1.5 * IQR_likes

df_cleaned_likes = df[(df['Likes'] >= lower_bound_likes) & (df['Likes'] <= upper_bound_likes)]
df_cleaned_likes.shape
```

```python
# Comments için aykırı değer tespiti
Q1_comments = df['Comments'].quantile(0.25) 
Q3_comments = df['Comments'].quantile(0.75)  
IQR_comments = Q3_comments - Q1_comments  

lower_bound_comments = Q1_comments - 1.5 * IQR_comments
upper_bound_comments = Q3_comments + 1.5 * IQR_comments

df_cleaned_comments = df[(df['Comments'] >= lower_bound_comments) & (df['Comments'] <= upper_bound_comments)]
df_cleaned_comments.shape
```

```python
outliers_views = df[(df['Views'] < lower_bound) | (df['Views'] > upper_bound)]
outliers_views
```

```python
outliers_likes = df[(df['Likes'] < lower_bound_likes) | (df['Likes'] > upper_bound_likes)]
outliers_likes
```

```python
outliers_comments = df[(df['Comments'] < lower_bound_comments) | (df['Comments'] > upper_bound_comments)]
outliers_comments
```

<h1>En Popüler Sanatçılar</h1>

```python
artist_popularity = df.groupby("Artist")[["Views", "Stream"]].sum()
artist_popularity = artist_popularity.sort_values("Views", ascending=False)
top_artists = artist_popularity.head(10)
top_artists
```

<h1>Beğeni-Oranlı Performans</h1>

```python
df["like_ratio"] = df["Likes"] / df["Views"]

# 2. Sanatçı başına ortalama beğeni oranı
artist_like_ratio = df.groupby("Artist")["like_ratio"].mean()

# 3. Beğeni oranına göre en yüksek sanatçılar
top_like_ratio_artists = artist_like_ratio.sort_values(ascending=False).head(10)
top_like_ratio_artists
```

<h1>Yorumlara Göre Sanatçı Etkileşimi</h1>

```python
# 1. Sanatçı başına toplam Yorum (Comments) ve İzlenme (Views) sayısı
artist_comments_views = df.groupby("Artist")[["Comments", "Views"]].sum()

correlation = artist_comments_views["Comments"].corr(artist_comments_views["Views"])
print(f"Comments ve Views arasındaki korelasyon: {correlation}")

# 3. Yorum sayısına göre en çok etkileşim alan sanatçılar
top_comment_artists = artist_comments_views.sort_values("Comments", ascending=False).head(10)
top_comment_artists
```

<h2>Yorum (Comments) ve izlenme (Views) sayıları arasında pozitif bir ilişki vardır (korelasyon: 0.43). Ancak bu ilişki çok güçlü değildir; yani yüksek izlenme sayıları her zaman yüksek yorum sayısı ile sonuçlanmayabilir.</h2>

<h1>Platform Performans Farkları</h1>

```python
platform_comparison = df.groupby("Artist")[["Views", "Stream"]].sum()

# 2. Performans farkını hesaplamak 
platform_comparison["Difference"] = platform_comparison["Stream"] - platform_comparison["Views"]

# 3. Hangi platformda daha popüler olduklarını belirlemek
platform_comparison["Preferred Platform"] = platform_comparison["Difference"].apply(
    lambda x: "Spotify" if x > 0 else "YouTube"
)

# 4. İlk 10 sanatçı
platform_comparison.head(10)
```

<h1>Kanal İstatistikleri</h1>

```python
kanal_verileri = df[['Channel', 'Views', 'Likes', 'Comments']]

kanal_istatistik = kanal_verileri.groupby('Channel').agg({'Views': 'mean', 
                                                          'Likes': 'mean', 
                                                          'Comments': 'mean'}).sort_values(by = [
                          "Views", "Likes", "Comments"], ascending = [False, False, False]).head(5).reset_index()
kanal_istatistik
```

```python
plt.figure(figsize=(18,8))

# Renkler 
bar_colors = ['#FF5733', '#33FF57', '#5733FF']

# İzlenme Grafiği
plt.subplot(1, 3, 1)
plt.bar(kanal_istatistik['Channel'], 
        kanal_istatistik['Views'], 
        color=bar_colors[0])
plt.xlabel('Kanal')
plt.ylabel('Toplam İzlenme')
plt.title('Toplam İzlenme')
plt.xticks(rotation=90, fontsize=15)

# Beğeni Grafiği
plt.subplot(1, 3, 2)
plt.bar(kanal_istatistik['Channel'], 
        kanal_istatistik['Likes'],
        color=bar_colors[1])
plt.xlabel('Kanal')
plt.ylabel('Toplam Beğeni')
plt.title('Toplam Beğeni')
plt.xticks(rotation=90, fontsize=15)

# Yorum Grafiği
plt.subplot(1, 3, 3)
plt.bar(kanal_istatistik['Channel'], 
        kanal_istatistik['Comments'], 
        color=bar_colors[2])
plt.xlabel('Kanal')
plt.ylabel('Toplam Yorum')
plt.title('Toplam Yorum')
plt.xticks(rotation=90, fontsize=15)

plt.tight_layout()
plt.show()
```

<h2>Wiz Khalifa Music ve LuisFonsiVEVO, milyonlarca görüntülenme ve beğeni ile öne çıkarken, Cocomelon izlenme almasına rağmen yorum almamış. Bu, Cocomelon'un daha geniş bir izleyici kitlesine hitap ettiğini ama daha düşük etkileşim oranlarına sahip olduğunu gösteriyor.</h2>

```python
labels = kanal_istatistik['Channel']
views = kanal_istatistik['Views']
likes = kanal_istatistik['Likes']
comments = kanal_istatistik['Comments']

# 3 adet pie chart ; 
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# İzlenme Pasta Grafiği ; 
axs[0].pie(views, labels=labels, autopct='%1.1f%%', startangle=140)
axs[0].set_title('İzlenme Dağılımı', fontsize=18)  # Başlığı daha büyük yapın

# Beğeni Pasta Grafiği ; 
axs[1].pie(likes, labels=labels, autopct='%1.1f%%', startangle=140)
axs[1].set_title('Beğeni Dağılımı', fontsize=18)  # Başlığı daha büyük yapın

# Yorum Pasta Grafiği ; 
axs[2].pie(comments, labels=labels, autopct='%1.1f%%', startangle=140)
axs[2].set_title('Yorum Dağılımı', fontsize=18)  # Başlığı daha büyük yapın


plt.suptitle('Kanal Bazında İzlenme, Beğeni ve Yorum Dağılımı', fontsize=25)

plt.tight_layout()
plt.show()
```
